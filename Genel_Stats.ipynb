```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import os
from datetime import datetime
import time
import io # StringIO için gerekli kütüphane

def scrape_fbref_tables_selenium():
    """
    FBref.com'daki Süper Lig istatistik tablolarını Selenium kullanarak çeker,
    BeautifulSoup ile ayrıştırır, çok seviyeli başlıkları düzleştirir ve
    her birini ayrı bir CSV dosyasına kaydeder.
    FutureWarning: Passing literal html to 'read_html' is deprecated düzeltmesi yapıldı.
    """
    
    url = "https://fbref.com/en/comps/26/Super-Lig-Stats#all_results2025-2026261"
    output_directory = "scraped_fbref_data" # Kazınan verilerin kaydedileceği klasör
    
    # Çıktı klasörünü kontrol et ve yoksa oluştur
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
        print(f"'{output_directory}' klasörü oluşturuldu.")

    # Kazınacak tabloların kullanıcı dostu isimleri ve HTML ID'leri
    table_configs = {
        "Super_Lig_Table_Overall": "results2025-2026261_overall",
        "Super_Lig_Table_Home_Away": "results2025-2026261_home_away",
        "Squad_Standard_Stats": "stats_squads_standard_for",
        "Opponent_Standard_Stats": "stats_squads_standard_against",
        "Squad_Goalkeeping": "stats_squads_keeper_for",
        "Opponent_Goalkeeping": "stats_squads_keeper_against",
        "Squad_Shooting": "stats_squads_shooting_for",
        "Opponent_Shooting": "stats_squads_shooting_against",
        "Squad_Playing_Time": "stats_squads_playing_time_for",
        "Opponent_Playing_Time": "stats_squads_playing_time_against",
        "Squad_Miscellaneous_Stats": "stats_squads_misc_for",
        "Opponent_Miscellaneous_Stats": "stats_squads_misc_against",
    }

    # Selenium WebDriver ayarları
    options = Options()
    options.add_argument("--headless")  # Tarayıcıyı arka planda çalıştır (sunucular için kritik)
    options.add_argument("--no-sandbox") # Linux konteynerleri için gerekli olabilir
    options.add_argument("--disable-dev-shm-usage") # Büyük dosyalar için /dev/shm kullanımını devre dışı bırak
    options.add_argument("start-maximized") # Tam ekran modunda başlat (headless olsa bile)
    options.add_argument("disable-infobars") # Bilgi çubuklarını devre dışı bırak
    options.add_argument("--disable-extensions") # Uzantıları devre dışı bırak
    options.add_argument("--disable-gpu") # GPU donanım hızlandırmasını devre dışı bırak
    options.add_argument(f'user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36')
    
    # WebDriver başlatma
    print("WebDriver başlatılıyor...")
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    print("WebDriver başarıyla başlatıldı.")

    try:
        # URL'ye git ve sayfanın yüklenmesini bekle
        print(f"URL'ye gidiliyor: {url}")
        driver.get(url)
        time.sleep(5) # Sayfanın tamamen yüklenmesi için bekle (gerektiğinde artırılabilir)
        print("Sayfa yüklendi. HTML içeriği çekiliyor...")

        # Tam HTML içeriğini al
        html_content = driver.page_source
        print("HTML içeriği başarıyla çekildi.")

        # BeautifulSoup ile HTML'i ayrıştırma
        soup = BeautifulSoup(html_content, 'lxml')
        print("HTML BeautifulSoup ile ayrıştırıldı.")

        # Her bir tabloyu ID'sine göre bulma ve kaydetme
        for friendly_name, table_id in table_configs.items():
            print(f"'{friendly_name}' tablosu aranıyor (ID: {table_id})...")
            
            # BeautifulSoup kullanarak belirli ID'ye sahip tabloyu bul
            table = soup.find('table', {'id': table_id})

            if table:
                try:
                    # >>> Düzeltme burada yapıldı: str(table) yerine io.StringIO(str(table)) kullanıldı <<<
                    dfs = pd.read_html(io.StringIO(str(table)), flavor='lxml')
                    
                    if dfs:
                        df = dfs[0] # İlk DataFrame'i al

                        # Çok seviyeli sütun başlıklarını düzleştirme
                        if isinstance(df.columns, pd.MultiIndex):
                            flattened_columns = []
                            for col_level_0, col_level_1 in df.columns:
                                if 'Unnamed' in col_level_0 or col_level_0 == "":
                                    if 'Unnamed' in col_level_1 or col_level_1 == "":
                                        flattened_columns.append(f"Column_{len(flattened_columns) + 1}")
                                    else:
                                        flattened_columns.append(col_level_1.strip())
                                else:
                                    flattened_columns.append(f"{col_level_0.strip()} - {col_level_1.strip()}")
                            df.columns = flattened_columns
                        else:
                            df.columns = [col.strip() for col in df.columns]

                        # Dosya adı için güncel tarih ekle
                        current_date = datetime.now().strftime("%Y-%m-%d")
                        filename = os.path.join(output_directory, f"{friendly_name}_{current_date}.csv")
                        
                        # DataFrame'i CSV olarak kaydet
                        df.to_csv(filename, index=False, encoding='utf-8')
                        print(f"'{friendly_name}' tablosu '{filename}' olarak başarıyla kaydedildi.")
                    else:
                        print(f"Uyarı: '{friendly_name}' (ID: {table_id}) için pandas.read_html DataFrame döndürmedi.")
                except Exception as e:
                    print(f"Hata: '{friendly_name}' tablosu pandas.read_html ile işlenirken bir sorun oluştu: {e}")
            else:
                print(f"Uyarı: '{friendly_name}' (ID: {table_id}) ID'sine sahip tablo BeautifulSoup tarafından bulunamadı.")
            
            time.sleep(2) # Her tablo işleme arasına kısa bir gecikme ekle

    except Exception as e:
        print(f"Genel bir hata oluştu: {e}")
    finally:
        # WebDriver'ı kapatmayı unutmayın
        print("WebDriver kapatılıyor...")
        driver.quit()
        print("WebDriver kapatıldı.")

if __name__ == "__main__":
    scrape_fbref_tables_selenium()
```
